```python
import sys, os

# Project root is the folder containing this notebook
project_root = os.path.abspath(".")
if project_root not in sys.path:
    sys.path.append(project_root)

# Output directory
output_dir = os.path.join(project_root, "output")

# Load config
import yaml
config_path = os.path.join(project_root, "config.yaml")
with open(config_path, 'r') as f:
    config = yaml.safe_load(f)

# Validate config
required_keys = ['qc_settings', 'valid_method_codes', 'valid_matrix_types', 'plot_colors', 'crm_line_colors', 'matrix_contexts', 'method_contexts']
for key in required_keys:
    if key not in config:
        raise ValueError(f"Missing required config key: {key}")

# Validate tolerances
qc_settings = config['qc_settings']
if not (0 < qc_settings['crm_low_tol'] < qc_settings['crm_high_tol'] < 200):
    raise ValueError("Invalid CRM tolerance range")

if not (0 < qc_settings['duplicate_rpd_tol'] < 100):
    raise ValueError("Invalid duplicate RPD tolerance")

if not (0 < qc_settings['duplicate_rpd_bdl_tol'] < 100):
    raise ValueError("Invalid duplicate BDL substitution RPD tolerance")

if not (0 < qc_settings['blank_tol_factor'] < 10):
    raise ValueError("Invalid blank tolerance factor")

from qc_engine.parser import load_metadata, load_qc_table
from qc_engine.column_namer import rename_analyte_columns
from qc_engine.classifier import classify_qc_table
from qc_engine.report_metadata import extract_all_identifiers

# ------------------------------------------------------------
# User Inputs (only update these)
# ------------------------------------------------------------
path = "G:/GIT/Python-LabQAQC/data/A25-15567Final.xlsx"   # <-- Update file here
lab_name = "Actlabs"                                      # <-- Update lab name here
meta_key = "Report Number"                                # <-- Update metadata key here
date_key = "Report Date"                                  # <-- Update date key here
qc_sheet = "QC"                                           # <-- Update Excel sheet name here

# QC interpretation settings
method_code = "AR-ICPMS"                                     # <-- Update assay method code here. See below for the list of options 
matrix_type = "vegetation"                                      # <-- Update sample type here. See below for a list of options

# QC tolerance settings (loaded from config, can override)
crm_low_tol = config['qc_settings']['crm_low_tol']                                        # CRM lower recovery tolerance (%), default = 80
crm_high_tol = config['qc_settings']['crm_high_tol']                                      # CRM upper recovery tolerance (%), default = 120
duplicate_rpd_tol = config['qc_settings']['duplicate_rpd_tol']                                  # Duplicate RPD tolerance (%), default = 30
duplicate_rpd_bdl_tol = config['qc_settings']['duplicate_rpd_bdl_tol']                            # Duplicate RPD tolerance for BDL substitution cases, default = 50
blank_tol_factor = config['qc_settings']['blank_tol_factor']                                    # Blank exceedance factor (e.g., 3× DL), default = 3

# BDL substitution settings
bdl_sub_rule = config['qc_settings']['bdl_sub_rule']                                     # options: "half", "sqrt2", "zero", "dl", or a numeric value

# QC interpretation settings options (from config)
valid_method_codes = config['valid_method_codes']
valid_matrix_types = config['valid_matrix_types']

print("Valid method_code options:", valid_method_codes)
print("Valid matrix_type options:", valid_matrix_types)
```

```python
from qc_engine.parser import load_metadata, load_qc_table
from qc_engine.column_namer import rename_analyte_columns
from qc_engine.classifier import classify_qc_table
import importlib
import qc_engine.duplicate_wide
importlib.reload(qc_engine.duplicate_wide)
from qc_engine.duplicate_wide import compute_duplicate_rpd

metadata = load_metadata(path)
df_qc = load_qc_table(path)

df_qc = rename_analyte_columns(df_qc, metadata)
df_qc = classify_qc_table(df_qc)

dup_results = compute_duplicate_rpd(
    df_qc,
    metadata,
    rpd_tolerance=duplicate_rpd_tol,
    bdl_substitution_rpd_tolerance=duplicate_rpd_bdl_tol,
)

dup_results.head()
```

```python
from qc_engine.crm_wide import compute_crm_recovery

crm_results = compute_crm_recovery(
    df_qc,
    metadata,
    low_tol=crm_low_tol,
    high_tol=crm_high_tol
)

crm_results.head()
```

```python
from qc_engine.blank_wide import compute_blank_qc

blank_results = compute_blank_qc(
    df_qc,
    metadata,
    tolerance_factor=blank_tol_factor,
    bdl_rule=bdl_sub_rule
)

blank_results.head()
```

```python
from qc_engine.qc_summary import build_qc_summary

summary = build_qc_summary(crm_results, dup_results, blank_results)
summary.head()
```

```python
# Pass crm_results directly — plot_crm_recovery() uses per-run data to draw one line per CRM
# crm_summary is still available from the summary table for reference
crm_summary = summary[["Analyte", "CRM_Recovery", "CRM_FinalStatus"]]

# Duplicate summary for plotting
dup_summary = summary[["Analyte", "Duplicate_RPD", "Duplicate_Status"]]

# Blank summary for plotting (use raw blank_results ONLY)
blank_plot_data = blank_results.rename(columns={"Status": "Blank_Status"})
```

```python
from qc_engine.qc_plots import plot_crm_recovery, plot_duplicate_rpd, plot_blank_levels

# Build certificate order from metadata
certificate_order = [metadata[i]["analyte"] for i in sorted(metadata.keys())]

# plot_crm_recovery takes crm_results (raw per-run data) to draw one line per CRM material
fig_crm = plot_crm_recovery(crm_results, certificate_order)
fig_dup = plot_duplicate_rpd(dup_summary, certificate_order)
fig_blank = plot_blank_levels(blank_plot_data, certificate_order)
```

```python
from qc_engine.report_export import save_qc_plot

file_name, metadata_name, report_date, names_match = extract_all_identifiers(
    path,
    metadata,
    meta_key=meta_key,
    date_key=date_key,
)

save_qc_plot(
    fig=fig_crm,
    lab_name=lab_name,
    metadata_name=metadata_name,
    report_date=report_date,
    plot_type="CRM Recovery",
    outdir=output_dir
)

save_qc_plot(
    fig=fig_dup,
    lab_name=lab_name,
    metadata_name=metadata_name,
    report_date=report_date,
    plot_type="Duplicate RPD",
    outdir=output_dir
)

save_qc_plot(
    fig=fig_blank,
    lab_name=lab_name,
    metadata_name=metadata_name,
    report_date=report_date,
    plot_type="Blank Levels",
    outdir=output_dir
)
```
